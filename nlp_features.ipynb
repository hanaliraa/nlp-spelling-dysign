```python
import os
import re
# import enchant.checker as enchant
import pandas as pd

def analyze_file(filename):
    # Read the file
    with open(filename, 'r') as f:
        text = f.read()
    
    # Compute the vocabulary length and complexity
    words = re.findall(r'\b\w+\b', text.lower())
    unique_words = set(words)
    vocab_length = len(unique_words)
    complexity = sum(1 for word in unique_words if words.count(word) > 1)
    
    # Compute the number of spelling errors
    # d = enchant.Dict("en_US")
    # errors = sum(1 for word in words if not d.check(word))
    
    return filename, vocab_length, complexity, None #, errors

# Set the path to the folder containing the text files
folder_path = "D:\OneDrive - Habib University\Kaavish\\nlp-spelling-dysign\data_transcibed"

# Create an empty dataframe to store the results
columns = ['Filename', 'Vocabulary Length', 'Complexity', 'Spelling Errors']
results_df = pd.DataFrame(columns=columns)

# Iterate over all the files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)
        filename, vocab_length, complexity, errors = analyze_file(file_path)
        results_df = results_df.append({
            'Filename': filename,
            'Vocabulary Length': vocab_length,
            'Complexity': complexity,
            'Spelling Errors': errors
        }, ignore_index=True)

# Export the dataframe to a CSV file
results_df.to_csv("results.csv", index=False)
```

```python
def analyse_text(sentences):
    vocab_lst = []
    complexity_lst = []
    for s in sentences:
        # Compute the vocabulary length and complexity
        words = re.findall(r'\b\w+\b', s.lower())
        unique_words = set(words)
        vocab_length = len(unique_words)
        complexity = sum(1 for word in unique_words if words.count(word) > 1)
        # Compute the number of spelling errors
        # d = enchant.Dict("en_US")
        # errors = sum(1 for word in words if not d.check(word))
        vocab_lst.append(vocab_length)
        complexity_lst.append(complexity)
    return vocab_lst, complexity_lst

df = pd.read_csv('data.csv')
sentences = list(df['sentence'])
v, c = analyse_text(sentences)
df['vocab_length'] = v
df['complexity'] = c

df.to_csv('data_v2.csv')
```
